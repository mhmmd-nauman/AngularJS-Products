```
def get_user_handles_v1(self, handle):
  """Given `screen_name` return dict from screen_name to a list of
  handles/friends dicts
  """
  handles = self.friends(handle)
  data =  []
  for handle in handles:
    mined = {
      'handle':handle.screen_name,
      'id':handle.id_str
    }
    data.append(mined)
  return data

a = get_user_handles_v1(api, "@mhmmd_nauman")
print(a) 
```

```
def get_tweet_histories(self, authors: Iterable[Author],
                        count=1000) -> Dict[Author, List[TweetData]]:
    """Given `screen_names` return dict from screen_name to a list of
    tweet_data dicts
    """
    return {author: self.get_tweet_history(author, count=count)
            for author in authors}
```

```
def show_histories(author_tweet_histories: Dict[Author, Iterable[TweetData]],
                   event_format=False) -> None:
    if event_format:
        writer = csv.writer(sys.stdout, delimiter=',')
        headers = ('unified_id', 'symbol', 'date', 'date_eastern',
                   'preannounce', 'trade_dir', 'author', 'tweet_body')
    else:
        writer = csv.writer(sys.stdout, delimiter='\t')
        headers = ('author', 'date', 'date_eastern', 'preannounce', 'trade_dir',
                   'new_tickers', 'tickers', 'tweet_body')
    writer.writerow(headers)
    for author, tweet_history in author_tweet_histories.items():
        for tweet_data in sorted(tweet_history, key=lambda x: x['date']):
            writer.writerow([tweet_data[_] for _ in headers])


def stay_subscribed(stock_stream: tweepy.Stream, author_ids: List[str],
                    at_screen_names: List[str]):
    """Subscribe and resubscribe on timeout errors. If resubscribing is
    attempted too often, Twitter will start sending error HTTP status_code
    420. So, code makes sure to respect min_resubscribe_wait"""
    min_resubscribe_wait = 120
    while True:
        subscription_time = time.time()
        try:
            stock_stream.filter(follow=author_ids)  # , async=True
        # except AttributeError as err:
        #     # workaround for something fixed in github
        #     if str(err) == "'NoneType' object has no attribute 'strip'":
        #         logger.warning('%s, re-subscribing', err)
        except socket.timeout as err:
            subscribed_timedelta = time.time() - subscription_time
            resubscribe_wait = round(
                max(min_resubscribe_wait - subscribed_timedelta, 0))
            logger.error('timeout error: %s, resubscribing after %s secs',
                         err, resubscribe_wait)
            notify_status((
                    '⚠ subscription timed out, re-subscribing '
                    'to {accounts} in {wait} seconds. Error: {err}'
                ).format(
                    accounts=','.join(at_screen_names),
                    wait=resubscribe_wait,
                    err=err),
                notifier=args.notify_status)
            time.sleep(resubscribe_wait)

            
```

# Scrape Tweets for Handles & Store In CSV

```
#userID = "@elonmusk"
#userID = "@CitronResearch"
#userID = "@Carl_C_Icahn"
userID = "@post_market"
tweets = api.user_timeline(screen_name=userID, 
                           # 200 is the maximum allowed count
                           count=200,
                           include_rts = False,
                           # Necessary to keep full_text 
                           # otherwise only the first 140 words are extracted
                            tweet_mode = 'extended'
                           )
```

```
all_tweets = []
all_tweets.extend(tweets)
oldest_id = tweets[-1].id
while True:
    tweets = api.user_timeline(screen_name=userID, 
                           # 200 is the maximum allowed count
                           count=200,
                           include_rts = False,
                           max_id = oldest_id - 1,
                           # Necessary to keep full_text 
                           # otherwise only the first 140 words are extracted
                           tweet_mode = 'extended'
                           )
    if len(tweets) == 0:
        break
    oldest_id = tweets[-1].id
    all_tweets.extend(tweets)
    print('N of tweets downloaded till now {}'.format(len(all_tweets)))
```
